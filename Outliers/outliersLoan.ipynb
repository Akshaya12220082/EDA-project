```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Income (USD)'].fillna(df['Income (USD)'].median(), inplace=True)
df['Income Stability'].fillna(df['Income Stability'].mode()[0], inplace=True)
df['Type of Employment'].fillna(df['Type of Employment'].mode()[0], inplace=True)
df['Current Loan Expenses (USD)'].fillna(df['Current Loan Expenses (USD)'].median(), inplace=True)
df['Dependents'].fillna(df['Dependents'].mean(), inplace=True)
df['Credit Score'].fillna(df['Credit Score'].median(), inplace=True)
df['Has Active Credit Card'].fillna(df['Has Active Credit Card'].mode()[0], inplace=True)
df['Property Age'].fillna(df['Property Age'].median(), inplace=True)
df['Property Location'].fillna(df['Property Location'].mode()[0], inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder

categorical_features = ['Gender', 'Income Stability', 'Profession', 'Type of Employment', 'Location', 'Expense Type 1', 'Expense Type 2', 'Has Active Credit Card', 'Property Type', 'Property Location', 'Co-Applicant']
numerical_features = ['Age', 'Income (USD)', 'Loan Amount Request (USD)', 'Current Loan Expenses (USD)', 'Dependents', 'Credit Score', 'No. of Defaults', 'Property Age']

# One-hot encode categorical variables
df_encoded = pd.get_dummies(df, columns=categorical_features)

# separating rows with and without missing values in 'Loan Sanction Amount (USD)'
train_df = df_encoded[df_encoded['Loan Sanction Amount (USD)'].notnull()]
test_df = df_encoded[df_encoded['Loan Sanction Amount (USD)'].isnull()]

# Defining features and target
X_train = train_df.drop(columns=['Loan Sanction Amount (USD)', 'Customer ID', 'Name'])
y_train = train_df['Loan Sanction Amount (USD)']
X_test = test_df.drop(columns=['Loan Sanction Amount (USD)', 'Customer ID', 'Name'])

# Training the regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting missing values
df.loc[df['Loan Sanction Amount (USD)'].isnull(), 'Loan Sanction Amount (USD)'] = model.predict(X_test)

print(df)
```

```python
numerical_features = [
    'Age', 'Income (USD)', 'Income Stability', 'Loan Amount Request (USD)', 
    'Current Loan Expenses (USD)', 'Dependents', 'Credit Score', 
    'No. of Defaults', 'Property Age', 'Property Price', 'Loan Sanction Amount (USD)'
]

for col in numerical_features:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# number of outliers
def count_outliers(data, columns):
    outlier_counts = {}
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
        outlier_counts[column] = outliers.shape[0]
    return outlier_counts

outliers = count_outliers(df, numerical_features)

for col, count in outliers.items():
    print(f'Number of outliers in {col}: {count}')
```

```python
import pandas as pd
import numpy as np
from scipy import stats

numerical_features = [
    'Income (USD)', 'Loan Amount Request (USD)', 
    'Current Loan Expenses (USD)', 'Dependents', 
    'No. of Defaults', 'Property Age', 'Property Price', 'Loan Sanction Amount (USD)'
]

# Z-score method on cleaned Data
z_scores_cleaned = np.abs(stats.zscore(df_no_outliers_z[numerical_features]))
threshold = 3

# Identifying rows with outliers
outliers_remaining = np.where(z_scores_cleaned > threshold)
rows_to_remove = np.unique(outliers_remaining[0])

# Removing the rows with outliers
df_no_outliers_z_final = df_no_outliers_z.drop(df_no_outliers_z.index[rows_to_remove])

# Verifying that the outliers have been removed
z_scores_final = np.abs(stats.zscore(df_no_outliers_z_final[numerical_features]))

# the number of outliers in the final cleaned dataset
outlier_counts_final = {}

for col, z_col in zip(numerical_features, z_scores_final.T):
    outlier_counts_final[col] = np.sum(z_col > threshold)

for col, count in outlier_counts_final.items():
    print(f'Number of outliers in {col} in the final cleaned dataset: {count}')
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Assuming df_final_cleaned is Dataset after the previous cleaning
numerical_features = [
    'Income (USD)', 'Loan Amount Request (USD)', 
    'Current Loan Expenses (USD)', 'Dependents', 
    'No. of Defaults', 'Property Age', 'Property Price', 'Loan Sanction Amount (USD)'
]

# Z-scores on the latest cleaned Data
z_scores_final = np.abs(stats.zscore(df_no_outliers_z[numerical_features]))
threshold = 3

# Initializing a boolean mask to identify rows with outliers
mask = np.zeros(df_no_outliers_z.shape[0], dtype=bool)

# Iterating through the columns to find rows with outliers
for col, z_col in zip(numerical_features, z_scores_final.T):
    mask |= (z_col > threshold)

# Filtering out rows with outliers
df_final_no_outliers = df_no_outliers_z[~mask]

# Verifying the number of outliers in the final cleaned dataset
z_scores_final_no_outliers = np.abs(stats.zscore(df_final_no_outliers[numerical_features]))

# Recalculating the number of outliers in the dataset after final cleaning
outlier_counts_final_no_outliers = {}

for col, z_col in zip(numerical_features, z_scores_final_no_outliers.T):
    outlier_counts_final_no_outliers[col] = np.sum(z_col > threshold)

for col, count in outlier_counts_final_no_outliers.items():
    print(f'Number of outliers in {col} in the final cleaned dataset: {count}')

print("Final DataFrame after removing outliers:")
print(df_final_no_outliers)
```
